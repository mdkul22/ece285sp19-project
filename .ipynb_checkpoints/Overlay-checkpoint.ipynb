```python
class ImgDataset(td.Dataset):
    def __init__(self, root_dir):
        super(ImgDataset,self).__init__()
        self.images_dir=root_dir
        self.files=os.listdir(self.images_dir)
        
    def __len__(self):
        return len(self.files)
    
    def __repr__(self):
        pass
    def __getitem__(self,idx):
        img_path=os.path.join(self.images_dir,self.files[idx])
        img=Image.open(img_path).convert('RGB')
        transform=tv.transforms.Compose([tv.transforms.Resize((228,228)),tv.transforms.ToTensor(),tv.transforms.Normalize([0.5]*3,[0.5]*3)])
        x=transform(img)
        return x
```

```python
train_set=ImgDataset(dataset_root_dir)
```

```python
def myimshow(image, ax=plt):
    image = image.to('cpu').detach().numpy()
    image = np.moveaxis(image, [0, 1, 2], [2, 0, 1])
    image = (image + 1) / 2
    image[image < 0] = 0
    image[image > 1] = 1
    h = ax.imshow(image)
    ax.axis('off')
    return h
```

```python
fig1,(ax1,ax2)=plt.subplots(ncols=2)
myimshow(train_set[10],ax1)
myimshow(train_set[12],ax2)
```
